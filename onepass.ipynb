```python
import pandas as pd
import numpy as np
import joblib

from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("dataset.csv")

y = df["label"].values

X_df = df.drop(columns=["label"])
```

```python
X_dict = X_df.to_dict(orient="records")
```

```python
vectorizer = DictVectorizer(sparse=False)
X_vectorized = vectorizer.fit_transform(X_dict)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_vectorized)
```

```python
model = SGDClassifier(
    loss="log",          
    learning_rate="optimal",
    max_iter=1,              
    tol=None,
    random_state=42
)
```

```python
classes = np.array([0, 1])
```

```python
for i in range(len(X_scaled)):
    model.partial_fit(
        X_scaled[i].reshape(1, -1),
        np.array([y[i]]),
        classes=classes
    )

print("âœ… One-pass online training completed")
```

```python
y_pred = model.predict(X_scaled)

print("\nðŸ“Š Accuracy:")
print(accuracy_score(y, y_pred))

print("\nðŸ“„ Classification Report:")
print(classification_report(y, y_pred))

print("\nðŸ§® Confusion Matrix:")
print(confusion_matrix(y, y_pred))
```
